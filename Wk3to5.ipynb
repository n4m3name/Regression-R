##### **2.1:** Parameter estimation

- To estimate model parameters, we calculate the sum of squares $S_{xx}$ and the sum of cross $S_{xy}$:
  $$
  S_{xx} = \sum x^2_i - n\bar{x}^2 \qquad S_{xy} = \sum x_iy_i - n\bar{x}\bar{y}
  $$
  - These are found in the R output:
  - 
- We calculate the following:
1. $\hat \beta_1 = \frac{S_{xy}}{S_{xx}}$
2. $\hat \beta_2 = \ol y - \hat \beta_1 \ol x$
- Note that the above are not given by the ANOVA table R output

##### **2.2:** Variance estimation

- Estimation of $\sigma^2$ is performed as follows:
  $$
  SS_T = \sum_1^n y^2_i - n \bar y^2 \qquad SS_{Res} = SS_T - \hat \beta_1 S_{xy} = SS_T - SS_R
  $$
  $$
  MS_{Res} = \frac{SS_{res}}{n-p} = \hat \sigma^2 = \frac{SS_{Res}}{n-2}
  $$
  - $p$: # parameters
  - $MS_{Res}$ is an *unbiased* estimator for $\sigma^2$
    - Expected value of estimation is equal to true value
- The solution is given in the R output for ANOVA above:

##### **2.3:** Hypothesis testing: $t$ test

- We test hypothesis that slope is a constant, i.e. $H_0: \beta_1 = \beta_{10}$
- $t$ statistic is same as Z statistic, safe uses estimator $s$ for sample standard deviation instead of the true value $\sigma$:
  $$
  Z_0 = \frac{\hat \beta_1 - \beta_{10}}{\sqrt{\sigma^2/S_{xx}}} \qquad t_0 = \frac{\hat \beta_1 - \beta_{10}}{\sqrt{MS_{Res}/S_{xx}}}
  $$
  - More generally:
  $$
  Z = \frac{\bar x - \mu}{\sqrt{\sigma^2/n}} \qquad t_0 = \frac{\bar x - \mu}{\sqrt{s^2/n}}
  $$
  - The distribution is $t_{n-2}$
  - Compares observed value of $t_0$ with upper $\alpha/2$ percentage point of the $t_{n-2}$ distribution, i.e. $t_{\alpha/2, n-2}$
    - Reject if $|t_0| > t_{\alpha/2, n-2}$
- We have $t_0 = -12.86$ in the above output, and critical value (not shown) $t_{0.025, 18} = 2.101$. We reject $H_0: \beta_1 = 0$, as is apparent due to the extremely small p-value `Pr(>|t|)`

##### **2.4:** Analysis of varinace

- We use the following:
  - $SS_{Res} = \sum_1^n (y_i - \hat y)^2$; Residual/model sum of squares
  - $SS_T = SS_R + SS_{Res}$, where $SS_R = \hat \beta_1 S_{xy}$
    - Follows that $df_T = df_R + df_{Res}$, and
      - $df_T = n-1$, one degree of freedom lost on constraint $\sum y_i - \bar y = 0$
        - i.e. sum of deviations from mean $y_i-\bar y$ can be freely chosen, safe last value which must equal to the negated sum of all previous
      - $df_R = 1$, since $SS_R$ determined entirely by $\hat \beta_1$
        - The calculation of $\hat \beta_0$ is based on $\hat \beta_1$, thus only the latter truly varies freely
      - $df_{Res}$ has $n-2$, since we estimate the parameters $\beta_i$ (one degree, see previous point) and have the constraints on residuals $\sum e_i = \sum(y_i - \hat y_i) = 0$ and $sum x_i e_i = 0$
- The analysis of variance F test will test $H_0: \beta_1 = 0$
  $$
  F_0 = \frac{SS_R/df_R}{SS_{Res}/df_{Res}} = \frac{SS_r/1}{SS_{Res}/(n-2)} = \frac{MS_R}{MS_{Res}}
  $$
  - Clearly follows $F_{1, n-2}$
  - We have $E(MS_{Res}) = \sigma^2$, $E(MS_R) = \sigma^2 + \beta_1^2 S_{xx}$
  - Indicates (?) that if the observed value of $F_0$ is large, likely that $\beta_1 \neq 0$
    - We use noncentrality parameter
      $$
      \lambda = \frac{\beta_1 S_{xx}}{\sigma^2}
      $$
  - To test, we compute $F_0$ and reject $H_0: \beta_1 = 0$ if $F_0 > F_{\alpha, 1, n-2}$
- We have following results from table:

#### **2.5:** Confidence intervals on $\beta_0$, $\beta_1$ and $\sigma^2$

#### **2.6:** Interval estimation of mean response

- We want to obtain a $100(1 - \alpha)$% CI on $E(y|x_0)$, using $\hat \mu_{y | x_0}$
  - Note that $\hat \mu_{y | x_0}$ is normally distributed, since is a linear combination of observations $y_i$
- Note the following:
  $$
  \hat \mu_{y | x_0} = \hat \beta_0 + \hat \beta_1 x_0
  $$
  $$
  V(\hat \mu_{y | x_0}) = V(\hat \beta_0 + \hat \beta_1 x_0) = \sigma^2\left[\frac{1}{n} + \frac{(x_0 - \bar x)^2}{S_{xx}}\right]
  $$
- A $100(1 - \alpha)$% CI is obtained as follows:
  $$
  \hat \mu_{y | x_0} \pm t_{\alpha/2, n-2}\sqrt{MS_{Res}\left(\frac{1}{n} + \frac{(x_0 - \bar x)^2}{S_{xx}}\right)}
  $$

````python
%%R

# Example 2.8: Regression through origin (shelf-stocking data)

# This example illustrates the following
# 1. the "lsfit" command
# 2. the "lmrob" command (for contaminated data)
# 3. how to fit a linear model through origin with "lsfit" and "lm"
# 4. how to retrieve residuals from a fit
# 5. how to compute the fitted value using the residuals
# 6. how to plot the residuals and generate a multi-plot graph
# 7. how to plot confidence band for the regression line
#
# Before running this code [1] install and then load package "robustbase" in R; this package
# is needed for the robust regression command "lmrob", and [2] create a .prn version of the
# data set for textbook Example 2.8, name the variables t and y, and put this data set in the 
# working directory of R on your computer.

library(robustbase)

### [1] Reading data into R and checking data
#
#   see the previous R script for file handling information.

data2=read.table(file="2-8.prn", header=TRUE)  
attach(data2)  # data2 is a table in R. This allows variables in the table to be used individually.
print(data2)   # see if there data points are correct
````

````python
%%R

### [2] Data analysis (with different models and R commands)

plot(t,y)
title("Shelf-stock data analysis")


fit1=lsfit(t,y)        # least-squares regression with intercept
writeLines("LSPRINT, FIT1 (LS W INTERCEPT, DEFAULT):")
ls.print(fit1)
abline(fit1,col="red") # add the ls line to the scatter plot
writeLines("---------------------------")
writeLines("LSPRINT, FIT2 (LS NO INTERCEPT, lsfit(t, y,intercept=FALSE)):")
fit2=lsfit(t,y,intercept=FALSE)  # least-squares regression without intercept
ls.print(fit2)
abline(fit2,col="blue") # add the ls line to the scatter plot
writeLines("---------------------------")
fit3=lm(y~t,data2)   # fit a linear model with intercept; same as lsfit
writeLines("ANOVA, FIT3 (LM W INTERCEPT, DEFAULT):")
print(anova(fit3))   # view ANOVA part of the results
writeLines("SUMMARY, FIT3:")
print(summary(fit3)) # see other results (t-tests, F-tests, R-squared) 
writeLines("---------------------------")


# plot confidence intervals for the regression line; "predict" command works 
# with fit generated by "lm" command
newx=data.frame(x=t)   # specify x values at which predictions about y are to be made
band=predict(fit3, newx, interval="confidence")   # predictions for the E(y|x) at x=t
print(band)            #view the three vectors in band. The default confidence level is 0.95.
lines(t,band[,2],lty=2, col="red")
lines(t,band[,3],lty=2, col="red")



fit4=lm(y~0+t,data2)   # fit a linear model without intercept
writeLines("ANOVA, FIT4 (LM NO INTERCEPT, lm(y~0+t, data2)):")
print(anova(fit4))   # view ANOVA part of the results
writeLines("Summary, Fit4:")
print(summary(fit4)) # see other results (t-tests, F-tests, R-squared) 
abline(fit4,col="black") # add the ls line to the scatter plot
writeLines("---------------------------")

# Robust estimation of linear models for contaminated data. Must install
# and load up the robust package first.
#
fit5=lmrob(y~t)  # robust estimation of linear model
writeLines("SUMMARY, FIT5, LMROB NO INTERCEPT (DEFAULT):")
print(summary(fit5))
abline(fit5,col="green") # add the ls line to the scatter plot
````

### **Problem 4: [Week 5, Sep 30-Oct 6]**

Do question 2.32 in the textbook.

Consider simple linear regression model
$$
y = \beta_0 + \beta_1x + \epsilon
$$
where the intercept $\beta_0$ is known
- (a) Find the least-squares estimator of $\beta_1$ for this model. Does this answer seem reasonable?
- (b) What is the variance of the slope $\hat \beta_1$ for the least squares estimator found in part a?
- (c) Find a $100(1-a)$ CI for $\beta_1$. Is this interval narrower than the estimator for the case where both slope and intercept are unknown?

Solution:

(a) We have the following LSE for $\beta_1$:
$$
S(\beta_1) = \sum_i (y_i - \hat y_i) = \sum_i (y_i - (\beta_0 + \beta_1 x_i))^2
$$
- Thus we find the minimum:
  $$
  \frac{dS}{d\beta_1} \big|_{\hat \beta_1} = 2 \sum_i (y_i - \beta_0 - \beta_1 x_i)' \cdot (y_i - \beta_0 - \beta_1 x_i) = -2 \sum_i x_i(y_i - \beta_0 - \beta_1 x_i) = 0
  $$
  $$
  \Rightarrow  \hat \beta_1 = \frac{ \sum_i x_i(y_i - \beta_0)}{\sum_i x_i^2}
  $$
- This solution is reasonable because it is the LSE for regression through the origin except with $\beta_1 \neq 0$:
  - No-intercept model: $y = \beta_1 x + \epsilon$
  - Leads to the following LSE for slope:
    $$
    \hat \beta_1 = \frac{\sum_i y_ix_i}{\sum_i x^2_i}
    $$
  - In our case, we have the model $y'_i = y_i - \beta_0 \rightarrow y'_i = \beta_1x + \epsilon$
  - Leads to the following LSE for slope:
    $$
    \hat \beta_1 = \frac{\sum_i y'_ix_i}{\sum_i x^2_i} = \frac{\sum_i x_i(y_i - \beta_0)}{\sum_i x^2_i}_{//}
    $$ 

(b) In order to find $V(\hat \beta_1)$, we plug in derived formula for $\hat \beta_1$:
$$
V(\hat \beta_0) = (\sum x_i^2)^{-2}\sum x^2_i V(y_i)
$$
Since $V(y_i) = \sigma^2$, we have
$$
V(\hat \beta_0) = \frac{\sigma^2}{\sum x^2_i} _{//}
$$

(c) The $100(1 - \alpha)$ CI is the same as that of the model through the origin, i.e.
$$
\beta_1 \pm t_{\alpha/2, n-1} \sqrt{\frac{MS_{Res}}{\sum x_i^2}}
$$
We can't sa whether it's narrower than the interval for a full model since we can't compare $\frac{MS_{Res}}{S_{xx}}$ to $\frac{MS_{Res}}{\sum x_i^2}$.

Clearly $\sum x_i^2 \geq S_{xx} = \sum(x_i - \bar x)^2$, but when $\beta_0$ is unknown $MS_{Res}$ ($\hat \sigma^2$) will be smaller (1 less degree of freedom, larger denominator)

##### **NOTES ON LSE (ESSENTIAL)**

- We have the following formula for sum of squared errors:
  $$
  S(\beta_0, \beta_1) = \sum_i \epsilon_i^2 = \sum_i (y_i - \hat y_i)^2 = \sum_i (y_i - (\beta_0 + \beta_1 x_i))^2
  $$
  - The formula gives the vertical distance between each given point $y_i$ and the corresponding point $\hat y_i$ on the regression line, unknown
  - Our goal is to minimize the distance between these points:
    - We can achieve this by squaring all distances (thus all positive), summing, and adjusting $\hat \beta_0$, $\hat \beta_1$ until the total sum is as small as possible
    - This is equivalent to finding the **minimum of the function**
    - Because our function is a parabola ($x_i^2$) with all positive terms, we are guaranteed that the point where the derivative is $0$ is a minimum
  - Thus in order to minimize the distance, we take the partial derivative of the SSE $S(\beta_0, \beta_1)$ in terms of the betas and simplify to obtain formulas for $\hat \beta_0$ and $\hat \beta_1$
  - Solving these formulas gives the regression line
- i.e.
  $$
  \frac{\delta S}{\delta \beta_i} = -2 \sum_i (y_i - \hat \beta_0 - \hat \beta_1 x_i) = 0 \quad (0)
  $$
- Solving gives the following:
  - Given:
  $$
  S_{xx} = \sum_i (x_i - \bar x)^2 \\
  S_{xy} = \sum_i y_i(x_i - \bar x)^2
  $$
  We have:
  $$
  \begin{aligned}
  &\hat \beta_1 = \frac{S_{xy}}{S_{xx}} = \sum_i c_iy_i, \quad c_i = (x_i - \bar x)/S_{xx} \quad &(1)\\
  &\hat \beta_0 = \bar y - \hat \beta_1 \bar x \quad &(2)
  \end{aligned}
  $$
- In order to find the expected value, variance of $\hat \beta_1$, we use $(1)$:
  $$
  \begin{aligned}
  E(\hat \beta_1) &= E(\sum_i c_i y_i) \\
  &= \sum_i c_i E(y_i) = \sum_i c_i (\beta_0 + \beta_1 x_i) \\
  &= \beta_0 \sum c_i + \beta_1 \sum c_i x_i
  \end{aligned}
  $$
    - Note here that we have $E(\epsilon_i) = 0$ by assumption, we have:
      - $\sum_i c_i = 0$
      - $\sum_i c_i x_i = 1$
    - Thus:
      $$
       E(\hat \beta_1) = \beta_0 \sum c_i + \beta_1 \sum c_i x_i = \beta_1
      $$
  - Thus the variance of $\hat \beta_1$:
    $$
    V(\hat \beta_1) = V(\sum c_i y_i) = \sum c_i^2 V(y_i)
    $$
    Since $V(y_i) = \sigma^2$ (assumed), we have
    $$
    V(\hat \beta_1) = \sigma^2 \sum c_i^2 = \frac{\sigma^2}{S_{xx}}
    $$
  - Also:
    $$
    V(\hat \beta_0) = V(\bar y - \hat \beta_1 \bar x) = V(\bar y) + \bar{x}^2 V(\hat \beta_1) - 2 \bar x Cov(\bar y, \hat \beta_1) 
    $$
    Since variance of $\bar y$ is $V(\bar y) = \sigma^2 /n$, covariance between $\bar y$ and $\hat \beta_1$ is $0$ (can be shown, see AI), we have
    $$
    V(\hat \beta_0) = \sigma^2 (\frac{1}{n} + \frac{\bar{x}^2}{S_{xx}})
    $$

##### Matrix Formation

The design matrix X has the form:

##### Matrix Operations
```r

C = solve(XTX)     # Finds (X'X)^-1
```
- C matrix contains variances and covariances of coefficient estimates
- C[2,2] is variance of β̂₁

##### Standard Error and Degrees of Freedom
```r
betahat = fit1$coefficients    # Gets coefficient estimates

df = length(y)-length(betahat) # df = n-p where p is number of parameters
```

##### Confidence Interval Calculation
```r
lower12 = betahat[2] - qt(0.975,df)*sigma*sqrt(C[2,2])
upper12 = betahat[2] + qt(0.975,df)*sigma*sqrt(C[2,2])
```

The formula used is:
$$
\hat{\beta}_1 \pm t_{\alpha/2,n-p}\sigma\sqrt{C_{22}}
$$

Where:
- $\hat{\beta}_1$ is the estimate of β₁
- $t_{\alpha/2,n-p}$ is the t-critical value
- σ is the residual standard error
- $C_{22}$ is the (2,2) element of $(X'X)^{-1}$

This manual calculation gives the same result as R's built-in confidence interval functions.

##### \+ **Explanation**

##### Confidence Interval for Mean Response
For a given x0, the CI for mean response E(y|x0) is:
$$\hat{y}(x_0) \pm t_{\alpha/2,n-p}\sqrt{MS_{Res}[\frac{1}{n} + \frac{(x_0-\bar{x})^2}{S_{xx}}]}$$

Where:
- x0 is the point where we want to estimate mean response
- MSRes is the mean square error
- n is sample size
- p is number of parameters
- t is the t-critical value

##### Prediction Interval
For a new observation at x0:
$$\hat{y}(x_0) \pm t_{\alpha/2,n-p}\sqrt{MS_{Res}[1 + \frac{1}{n} + \frac{(x_0-\bar{x})^2}{S_{xx}}]}$$

##### Key Differences
1. Prediction interval is wider than CI because:
   - CI estimates mean response
   - PI accounts for both model uncertainty and individual observation variability
   - Note the extra "1" under the square root for PI

2. In R code:
```r
predict(model, newdata, interval="confidence")  # for CI
predict(model, newdata, interval="prediction")  # for PI
```

The wider prediction interval reflects the additional uncertainty in predicting an individual observation versus estimating a mean response.


```r

```
This gives the variance multiplier for the confidence interval.

##### Step 3: Confidence Interval for Mean Response

Uses formula: $$\hat{y}(x_0) \pm t_{\alpha/2,n-p}\sigma\sqrt{x_0'(X'X)^{-1}x_0}$$

Uses formula: $$\hat{y}(x_0) \pm t_{\alpha/2,n-p}\sigma\sqrt{1 + x_0'(X'X)^{-1}x_0}$$

The key difference between confidence and prediction intervals is:

##### \+ **Explanation:**

##### Multicollinearity

Multicollinearity occurs when predictor variables (x1, x2) in a regression model are correlated with each other. This can cause problems because:
- Makes it difficult to determine individual effects of predictors
- Increases variance of coefficient estimates
- Makes model unstable (coefficients sensitive to small changes)

##### VIF (Variance Inflation Factor)

The VIF function calculates how much the variance of a coefficient is inflated due to correlation with other predictors:

$$VIF_j = \frac{1}{1-R_j^2}$$

Where:
- R_j² is the R² from regressing xj on all other predictors
- VIF = 1 means no correlation with other predictors
- VIF > 5 or 10 indicates problematic multicollinearity

##### In This Code
```r
vif(fit1)  # Calculates VIF for x1 (cases) and x2 (distance)
```

- Cases and distance are not highly correlated
- Their coefficient estimates are reliable
- No remedial action needed for multicollinearity

This matches intuition - number of cases and delivery distance could vary independently of each other in delivery routes.

    - n = number of observations
    - k = number of parameters in full model (including intercept)
3. Use to obtain the following p-value:
   $$
   p = (F_{DF_{reduced}, DF_{full}} > F^*)
   $$
